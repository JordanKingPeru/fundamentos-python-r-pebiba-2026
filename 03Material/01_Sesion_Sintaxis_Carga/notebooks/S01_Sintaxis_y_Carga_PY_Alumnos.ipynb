### 💡 Pro-Tip: list comprehensions

Una vez te sientas cómodo con el `for`, puedes condensar lecturas simples en **list comprehensions**:

```python
# Estas dos formas son equivalentes:
cuadrados = []
for x in [1, 2, 3, 4]:
    cuadrados.append(x * x)

cuadrados = [x * x for x in [1, 2, 3, 4]]   # ← list comprehension
```

Las verás mucho en código Python real. Por ahora basta con que las **leas y entiendas**.

*En R el equivalente es `sapply(lista, función)` o `purrr::map(lista, función)`. Lo verás en la Sesión 02.*
