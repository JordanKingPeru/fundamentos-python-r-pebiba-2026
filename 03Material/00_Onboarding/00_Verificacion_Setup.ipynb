## ✅ Test 3: Carga y manipulación básica de DataFrame

```python
# Crear DataFrame mini para verificar operaciones
df = pd.DataFrame({
    'region': ['Lima', 'Cusco', 'Lima', 'Arequipa', 'Cusco'],
    'ventas': [1000, 500, 1500, 800, 700],
    'fecha': pd.to_datetime(['2026-01-15', '2026-02-20', '2026-03-10', '2026-04-05', '2026-05-12'])
})

print(df)
print(f'\nShape: {df.shape}')

# Operación básica: agrupar y agregar
resumen = df.groupby('region', as_index=False)['ventas'].sum()
print(f'\nResumen por región:\n{resumen}')

assert len(resumen) == 3, '3 regiones esperadas'
print('\nOK Manipulación de DataFrame funciona')
```

## ✅ Test 4: Visualización

```python
# Generar gráfico simple
fig, ax = plt.subplots(figsize=(8, 4))
resumen.plot(kind='bar', x='region', y='ventas',
             ax=ax, color='#8A1525', legend=False)
ax.set_title('Test de visualización - PEBIBA XVIII', fontweight='bold')
ax.set_ylabel('Ventas (PEN)')
plt.tight_layout()
plt.show()

print('OK Visualización funciona — viste el gráfico arriba')
```

## ✅ Test 5: Carga del dataset del curso

Verificar que el dataset Acme Corp está disponible y se carga correctamente.
